## –ó–∞–¥–∞–Ω–∏–µ 3

–í—ã–≥—Ä—É–∑–∏ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `posts` –≤ —Ñ–æ—Ä–º–∞—Ç–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ __–≤—Å–µ –∑–∞–ø–∏—Å–∏__ —Å–æ–æ–±—â–µ—Å—Ç–≤–∞ Skyeng: https://vk.com/skyeng, 


```python
#—Å–æ–±–µ—Ä–µ–º –≤—Å–µ –∑–∞–ø–∏—Å–∏ —Å–æ–æ–±—â–µ—Å—Ç–≤–∞ 
import time # –¥–ª—è sleep

offset = 0
max_count = 100
posts = pd.DataFrame()

while True:      # –ü–æ–≤—Ç–æ—Ä—è—Ç—å —Ü–∏–∫–ª –ø–æ–∫–∞ –µ–≥–æ –Ω–µ –æ—Å—Ç–∞–Ω–æ–≤—è—Ç –ø–æ —É—Å–ª–æ–≤–∏—é –Ω–∏–∂–µ
    
    url = 'https://api.vk.com/method/wall.get'
    params = {
        'access_token': token,
        'v':'5.89',
        'domain': 'skyeng',
        'count': max_count,
        'offset': offset
    }
    
    req = requests.get(url=url, 
                       params=params)
    res = req.json()
    
    if len(res['response']['items']) != 0:      # –ü—Ä–æ–≤–µ—Ä—è–µ–º –≤–µ—Ä–Ω—É–ª–∏—Å—å –ª–∏ –Ω–∞–º —Ö–æ—Ç—å –∫–∞–∫–∏–µ-–Ω–∏–±—É–¥—å –∑–∞–ø–∏—Å–∏
        posts = pd.concat([posts, pd.DataFrame(res['response']['items'])]) 
    else:
        break                                   # –ï—Å–ª–∏ –Ω–∞–º –Ω–∏—á–µ–≥–æ –Ω–µ –≤–µ—Ä–Ω—É–ª–æ—Å—å, —Ç–æ –ø—Ä–µ—Ä–≤–∞—Ç—å —Ü–∏–∫–ª
    print('–£–∂–µ —Å–∫–∞—á–∞–Ω–æ', offset, '–∑–∞–ø–∏—Å–µ–π')
    offset = offset + max_count                 # –°–º–µ—â–∞–µ–º offset –Ω–∞ —á–∏—Å–ª–æ –≤–æ–∑–≤—Ä–∞—â–µ–Ω–Ω—ã—Ö –∑–∞–ø–∏—Å–µ–π
    time.sleep(0.5)                             # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º —á–∞—Å—Ç–æ—Ç—É –∑–∞–ø—Ä–æ—Å–æ–≤ –¥–æ –æ–¥–Ω–æ–≥–æ –∑–∞–ø—Ä–æ—Å–∞ –≤ –ø–æ–ª—Å–µ–∫—É–Ω–¥—ã
```

## –ó–∞–¥–∞–Ω–∏–µ 4

–í—ã–≥—Ä—É–∑–∏ 50 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–æ—Å—Ç–æ–≤. –ú–µ—Ç—Ä–∏–∫–æ–π –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –º—ã —Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ª–∞–π–∫–æ–≤, —Ä–µ–ø–æ—Å—Ç–æ–≤ –∏ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤.


```python
# –≤—ã—Ç–∞—â–∏–º –∏–∑ —Å–ª–æ–≤–∞—Ä–µ–π —Å—Ç–æ–ª–±—Ü–æ–≤ comments, likes, reposts —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
posts_copy = posts.copy() #—Å–æ—Ö—Ä–∞–Ω–∏–º —Å—ã—Ä—ã–µ –¥–∞–Ω–Ω—ã–µ
def get_value(name_column):
    posts_copy[name_column + '_count'] = posts_copy[name_column]
    for i in range(0, len(posts_copy)):
        posts_copy[name_column + '_count'].iloc[i] = posts_copy[name_column].iloc[i]['count']


    
```

```python
column_name_list = ['comments', 'likes', 'reposts']
for name in column_name_list:
    get_value(name)
    
```

```python
posts_copy.head()
```

```python
# –∑–∞–ø–æ–ª–Ω—è–µ–º –¥–∞–Ω—ã–º–∏ —á–µ—Ä–µ–∑ lambda —Ñ—É–Ω–∫—Ü–∏—é
posts['comments_count'] = posts['comments'].apply(lambda x: x.get('count'))
posts['likes_count'] = posts['likes'].apply(lambda x: x.get('count'))
posts['reposts_count'] = posts['reposts'].apply(lambda x: x.get('count'))
posts.head()
```

```python
# –∑–∞–ø–æ–ª–Ω—è–µ–º –¥–∞–Ω–Ω—ã–º–∏ —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–æ–π –ø–æ –≤—Å–µ–º –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—è–º
posts['activity'] = posts[['comments_count','likes_count', 'reposts_count']].sum(axis=1)
```

```python
#–ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞—Ç—É –ø–æ—Å—Ç–∞ –∫ –ø—Ä–∏–≤—ã—á–Ω–æ–º—É —Ñ–æ—Ä–º–∞—Ç—É
from datetime import datetime
posts['date'] = posts['date'].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
#–≤—ã–±–∏—Ä–∞–µ–º 50 —Å–∞–º—ãx –ø–æ–ø—É–ª—è—Ä–Ω—ãx 
posts_top50 = posts.sort_values(by = 'activity', ascending = False).reset_index().head(50)
posts_top50.head()
```

```python
#—Å–ø–∏—Å–æ–∫ id
list_posts_id = posts_id_top50['id'].tolist()
```

```python
#—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Ç–µ—Å—Ç–æ–≤—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ —Ä–∞–±–æ—Ç—ã —Ü–∏–∫–ª–∞
posts_id_top2 = posts_id_top50.iloc[0:11]
list_posts_id2 = posts_id_top2['id'].tolist() #–º–∞–ª–µ–Ω—å–∫–∏–π —Å–ø–∏—Å–æ–∫, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–≥–æ–¥–∏—Ç—Å—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Ü–∏–∫–ª–∞
posts_id_top2['comments_count'].sum()
```

```python
#—Å–æ–±–µ—Ä–µ–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ—Å—Ç–∞ –≤—Å–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –∏ –ª–∞–π–∫–∏ –∫ –Ω–∏–º

owner_id = str(posts['owner_id'].iloc[0])
offset = 0
max_count_comments = 100
comments = pd.DataFrame(columns = {'post_id'})

i = 0

while True:
    periods=0
    url = 'https://api.vk.com/method/wall.getComments'

    params = {'access_token': token,
                'v':'5.89',
                'owner_id': owner_id,
                'post_id': list_posts_id[i],
                'need_likes': 1, 
                'offset': offset,
                'count': max_count,
                'sort': 'desc',
                'preview_length': 0 #–Ω–µ –æ–±—Ä–µ–∑–∞–µ–º –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏
            }
    req = requests.get(url, params)
    res = req.json()
     
    

    if len(res['response']['items']) != 0: 
        df = pd.DataFrame(res['response']['items'])
        comments = pd.concat([comments, df])

        comments['post_id'] = comments['post_id'].shift(periods=periods) #–¥–æ–±–∞–≤–∏–º post_id —Å–æ —Å–º–µ—â–µ–Ω–∏–µ–º 
        comments['post_id'] = comments['post_id'].fillna(str(list_posts_id[i]))
        offset = offset + max_count_comments                 # –°–º–µ—â–∞–µ–º offset –Ω–∞ —á–∏—Å–ª–æ –≤–æ–∑–≤—Ä–∞—â–µ–Ω–Ω—ã—Ö –∑–∞–ø–∏—Å–µ–π
        periods = periods + len(df) #—Å–º–µ—â–µ–Ω–∏–µ –¥–ª—è poat_id
        time.sleep(0.5)
    else:
        if i < len(list_posts_id)-1:
            offset = 0
            periods = 0
            i +=1

        else:
            break
    
    
```
